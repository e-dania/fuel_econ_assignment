```python
colors = ['#cd7660', '#ffffff', '#4e899b']
custom_cmap = LinearSegmentedColormap.from_list("custom_heatmap", colors)
df_numeric = df.select_dtypes(include=np.number)
correlation = df_numeric.corr()
sns.heatmap(correlation, cmap=custom_cmap, center=0)
plt.show()
```

The visualization looking at it from the point of CO2,
GHG score, FE score, comb, UHighway, highway, Ucity, city are negatively correlated to CO2, which means CO2 inversely related to all of them

While pv2, display and cylinders are positively correlated to C02, which means that CO2 is directly correlated to them.

Id, Year and pv4 are close to 0 on the correlation bar (C bar) which means that they don't really have any effect on the CO2 score.

What i can understand from all this is that vehicles wih more cylinders, pv2, displays emit more CO2.
